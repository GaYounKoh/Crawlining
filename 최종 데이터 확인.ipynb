```python
발생dir = glob('정형철교수님/*/')[-1]
files = glob(발생dir+'/*')
```

```python
from tqdm import tqdm
```

```python
for i in tqdm(files):
    dat = pd.read_csv(i, encoding = 'cp949', header = None).iloc[:,:-1]
    if dat.isnull().sum().sum() > 0:
        print(i)
```
