```python
def myinfo(df_):
    a_ = df_.columns
    b_ = df_.shape
    c_ = df_.info()
    d_ = df_.head()
    e_ = df_.tail()
    
    return a_, b_, c_, d_, e_
```

```python
from tqdm import tqdm
```
