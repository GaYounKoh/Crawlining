```python
for i in range(질병.shape[1]):
    질병.iloc[:,i] = list(map(lambda x: np.nan if x == '-' else x, 질병.iloc[:,i])) # 질병.iloc[:,i].map({'-':np.nan})
```

```python
new질병 = 질병.copy()
```

```python
# new질병.년도 = new질병.년도.map(int)
# new질병.월 = new질병.월.map(int)

# new질병 = new질병.set_index(['년도', '월'])

idxli = []
for i in range(2001, 2023):
    if i == 2022:
        idxli.extend([f'{i}0{j}' if j < 10 else f'{i}{j}' for j in range(1, 9)])
    else:
        idxli.extend([f'{i}0{j}' if j < 10 else f'{i}{j}' for j in range(1, 13)])
        
## index를 int로 바꿔서 넣으면 그림 그릴 때 xticklabels가 연속되는 숫자로 이어져버림.
new질병.index = idxli
new질병.drop(['년도', '월'], axis = 1, inplace = True)
```

```python
new질병4exmp = new질병[['에볼라', '수두', '유행성이하선염', 'A형간염', '카바페넴', 'C형간염', '쯔쯔가무시', '지카바이러스']].iloc[-30:, :]
```
